```r
mod_tbl <- readxl::read_xlsx("../github/Fetal_metaQTL/data/manuscript/ST6.xlsx", sheet = "ST6-1-Module-Genes")
big_enrich_tbl <- readxl::read_xlsx("../github/Fetal_metaQTL/data/manuscript/ST6.xlsx", sheet = "ST6-2-All-Enrichment") %>%
                    mutate(fdr = p.adjust(p, "fdr")) %>%
                    mutate(sig = if_else(fdr <= 0.1, 1, 0))
```

```r
unique(big_enrich_tbl$enrich_method)
unique(big_enrich_tbl$enrich_group)
```

```r
big_enrich_tbl %>%
    filter(enrich_method == 'cell_type_ORA') %>%
    group_by(enrich_group) %>%
    arrange(fdr) %>%
    slice_head(n = 3) %>%
    ungroup
```

```r
new_to_old <- function(new_mod_name) {
    return (mod_tbl %>% filter(assigned_module == new_mod_name) %>% pull(classic_name) %>% unique)
}
```

```r
big_enrich_tbl_top <- big_enrich_tbl %>%
    filter(enrich_method == 'cell_type_ORA') %>%
    group_by(enrich_group) %>%
    arrange(fdr) %>%
    slice_head(n = 3) %>%
    ungroup
```

```r
ggplot_list <- list()
```

```r
for (i in 1:nrow(big_enrich_tbl_top)) {
    new_mod <- big_enrich_tbl_top$module[i]
    module <- new_to_old(new_mod)
    one_mod_me_tbl <- tibble(subject = rownames(full_me_tbl),
                         eigengene = full_me_tbl[[module]])

eigen_tbl <- inner_join(one_mod_me_tbl, meta_df, by = c("subject" = "Subject")) %>%
#                 mutate(pos_age = 1+age) %>%
#                 mutate(bins = cut(pos_age, breaks = 20)) %>%
#                 mutate(bins = as.numeric(bins)) %>%
                filter(trimester != 3)
#                 mutate(pcw_age = pos_age*40)

age_plot <- ggplot(eigen_tbl, aes(x = pcw, y = eigengene)) + 
    geom_point() + 
    geom_smooth(method = "lm") +
    theme(panel.grid.major = element_blank(), 
          panel.grid.minor = element_blank(),
          panel.background = element_blank(), 
          axis.line = element_line(colour = "black"), 
          text = element_text(family="Helvetica")) +
    ggtitle(paste0(big_enrich_tbl_top[i,'module'], "-", big_enrich_tbl_top[i,'enrich_group']))
    
  ggplot_list[[i]] <- age_plot

}
```

```r
# install.packages("gridExtra")
library(gridExtra)
```

```r
options(repr.plot.height = 40, repr.plot.width = 18)
# Combine the ggplots into a facet grid
facet_grid <- grid.arrange(grobs = ggplot_list, nrow = 9)

# Display the facet grid
print(facet_grid)
```
