```r
subject_sex <- c()
subject_tri <- c()

for (i in 1:ncol(df1)) {
    sample <- colnames(df1)[i]
    # sex
    subject_sex[i] <- metadata[metadata$Subject == sample, "inferSex"]
    # age, PCW, trimester
    pcw <- metadata[metadata$Subject == sample, "Age"]*365/7+40 
    if (pcw < 13){
        subject_tri[i] <- 1
    }
    if (13 <= pcw & pcw < 28){
        subject_tri[i] <- 2
    }
    if (pcw >= 28){
        subject_tri[i] <- 3
    }
}

subject_sex <- factor(subject_sex)
subject_tri <- factor(subject_tri)
```
